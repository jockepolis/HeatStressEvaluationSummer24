```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the specified path
file_path = '/Users/user/Desktop/Sommarjobb24/HeatStressEvaluation/Data/MergedData/TheFertilityData.csv'
df = pd.read_csv(file_path)

# Drop NaN values based on NINS
df = df.dropna(subset=['NINS'])

# Plot the distribution of NINS to identify outliers
plt.figure(figsize=(10, 6))
plt.boxplot(df['NINS'], vert=False)
plt.xlabel('Number of Inseminations (NINS)')
plt.title('Boxplot of NINS')
plt.show()

# Calculate statistical measures for NINS
nins_desc = df['NINS'].describe()
print(nins_desc)

# Identify outliers using IQR method
Q1 = df['NINS'].quantile(0.25)
Q3 = df['NINS'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['NINS'] < (Q1 - 1.5 * IQR)) | (df['NINS'] > (Q3 + 1.5 * IQR))]
print("Outliers based on IQR method:")
print(outliers[['SE_Number', 'NINS']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the specified path
file_path = '/Users/user/Desktop/Sommarjobb24/HeatStressEvaluation/Data/MergedData/TheFertilityData.csv'
df = pd.read_csv(file_path)

# Filter out negative values
df = df[(df['CFI'] >= 0) & (df['CLI'] >= 0) & (df['FLI'] >= 0) & (df['GL'] >= 0)]

# Define a function to analyze a variable
def analyze_variable(df, column_name, title):
    # Plot the distribution to identify outliers
    plt.figure(figsize=(10, 6))
    plt.boxplot(df[column_name], vert=False)
    plt.xlabel(title)
    plt.title(f'Boxplot of {title}')
    plt.show()

    # Calculate statistical measures
    desc = df[column_name].describe()
    print(desc)

    # Identify outliers using IQR method
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    outliers = df[(df[column_name] < (Q1 - 1.5 * IQR)) | (df[column_name] > (Q3 + 1.5 * IQR))]
    print(f"Outliers based on IQR method for {title}:")
    print(outliers[['SE_Number', 'LactationNumber', column_name]].drop_duplicates())

# Analyze CFI (Calving to First Insemination Interval)
analyze_variable(df, 'CFI', 'Calving to First Insemination Interval (CFI)')

# Analyze CLI (Calving to Last Insemination Interval)
analyze_variable(df, 'CLI', 'Calving to Last Insemination Interval (CLI)')

# Analyze FLI (First to Last Insemination Interval)
analyze_variable(df, 'FLI', 'First to Last Insemination Interval (FLI)')

# Analyze GL (Gestation Length)
analyze_variable(df, 'GL', 'Gestation Length (GL)')
```
